# FILTROS
A la hora de filtrar tenemos dos operadores logicos: 
**& (and) y  | (or)**

Para hacer uso de estos operadores tenemos que tener dos sentencias logicas con alguno de los operadores que ya vimos **(>, <, >=, <=, ==)**

Si queremos utilizar el **&** o el **|** tenemos que encerrar las sentencias entre parentesis. 

`(sentencia) & (sentencia) `


Podemos filtrar los presidentes por fecha

````python
from datetime import datetime
fecha_desde = datetime(1798,1,1)
fecha_hasta = datetime(1805,1,1)
df_filtrado = df[(df.start >= fecha_desde) & (df.start <= fecha_hasta)]
df[(df.start>= pd.to_datetime('1798-01-01')) & (df.start<= pd.to_datetime('1805-01-01'))] 

#diferencia = fecha_hasta - fecha_desde
#print(diferencia)
````

# FUNCIONES EXTRAS

#### Contar valores repetidos

````python
cant_mandatos = df.president.value_counts() # Por dault lo ordena de mayor cantidad a menor
````

Como podriamos mostrar el nombre de los presidentes que tuvieron mas de un mandato?

````python
cant_mandatos[df.president.value_counts() > 1]
````

Mostramos solo el nombre del presidente accediendo al indice y al primer valor con los corchetes [0]

````python
cant_mandatos[df.president.value_counts() > 1].index[0]
````

Mostramos la cantidad de presidencia pero de los valores por lo tanto no hay que acceder al indice solo a la ubicacion

````python
cant_mandatos[df.president.value_counts() > 1][0]
````

````python
df.president.value_counts().index[0] #Con el .index vemos los nombres de los presidentes, que es lo que estamos contando
````
