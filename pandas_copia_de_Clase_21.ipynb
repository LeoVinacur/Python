# Series

Se usan para modelar estrcturas 1D similar a las listas en Python aunque tienen mayor numero de bits incluyendo indice y nombre. 

Pueden albergar strings, floats, booleans

ejemplo.jpg

```python
import pandas as pd
# Pandas series
numeros = range(50, 70, 2)
numeros_serie = pd.Series(numeros)
print(numeros_serie)
```

```python
# Accediendo al valor asociado al indice 2 de la serie
print(numeros_serie[2])
```

```python
# Índices
print(numeros_serie.index)
```

```python
# Valores
print(numeros_serie.values)
```

```python
# Se modifican los índices
numeros_en_texto = ['primero','segundo','tercero','cuarto','quinto','sexto','séptimo','octavo','noveno','décimo']
numeros_serie = pd.Series(numeros, index = numeros_en_texto)
numeros_serie
```

```python
df_nan = numeros_serie_porcion1 + numeros_serie_porcion2
print(numeros_serie_porcion1 + numeros_serie_porcion2)
```
